```python
#combine chicago and new orleans listings dataset
combined = pd.concat([chicago, new_orleans], axis=0, ignore_index=True)
```

```python
# Missing Values/Null Values Count
print(combined.isnull().sum())
```

```python
# Write your code to make your dataset analysis ready.
# Let's Start with data cleaning.
# Remove duplicate rows from the hotel_booking DataFrame, modifying it in place
combined.drop_duplicates(inplace=True)
```

```python
#drop 'neighbourhood_group' column
combined.drop(columns=['neighbourhood_group'], inplace=True)
```

```python
#fill missing values in 'host_name' with 'no name'
combined['host_name'].fillna('No Name', inplace=True)
```

```python
#fill missing values in 'last_reviews' with 'no review' and 'reviews_per_month' with '0'
combined['last_review'].fillna('No Review', inplace=True)
combined['reviews_per_month'].fillna(0,inplace=True)
```

```python
# Convert 'last_review' to datetime
pd.to_datetime(combined['last_review'], errors='coerce')
```

```python
#fill missing values in 'license' and 'not provide'
combined['license'].fillna('not provided', inplace=True)
```

```python
#fill missings values in 'price' with mean
combined['price'].fillna(combined['price'].mean(), inplace=True)
```

```python
print(combined.isnull().sum())
```

```python
combined.head()
```

```python
combined.info()
```

```python
# Create 'price_category' column
def categorize_price(price):
    if price < 75:
        return 'Low'
    elif 75 <= price <= 150:
        return 'Medium'
    else:
        return 'High'

combined['price_category'] = combined['price'].apply(categorize_price)
```

```python
# Create 'review_activity' column based on reviews_per_month
def categorize_reviews(rpm):
    if rpm == 0:
        return 'Inactive'
    elif 0 < rpm <= 1:
        return 'Low'
    elif 1 < rpm <= 3:
        return 'Medium'
    else:
        return 'High'

combined['review_activity'] = combined['reviews_per_month'].apply(categorize_reviews)
```

```python
# Create 'host_type' column
combined['host_type'] = combined['calculated_host_listings_count'].apply(lambda x: 'Individual' if x == 1 else 'Business')
```

```python
combined.to_csv("cleaned_combined.csv", index=False)
```
